```python
import pandas as pd

# Step 1: CSV load karo
df = pd.read_csv("data-export (1).csv")

# Step 2: Pehli row ko header banao
df.columns = df.iloc[0]  # first row ko header banado
df = df[1:]               # first row remove kar do
df.reset_index(drop=True, inplace=True)
df.columns = ["channel group", "DateHour",						
"Users","Sessions","Engaged sessions","Average engagement time per session","Engaged sessions per user","Events per session",		
"Engagement rate","Event count"]				

df.head()
# Step 3: Check karo
df.head()


df["DateHour"] = pd.to_datetime(df["DateHour"], format="%Y%m%d%H", errors="coerce")#20240416 is next formate2024-04-16 23:00:00
df.head()



```

```python
df.info()
```

```python
numeric_cols = df.columns.drop(["channel group","DateHour"])
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric,errors='coerce')
df['Hour'] = df["DateHour"].dt.hour#alg se colums banayega.dt.year → साल

# .dt.month → 

# .dt.day 

# .dt.hour

# .dt.minute
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

```

```python
sns.set(style='whitegrid')#Seaborn style set karna (background ko white grid banane ke liye
plt.figure(figsize=(10,5))#width 10 height 5

df.groupby("DateHour")[["Sessions", "Users"]].sum().plot(ax=plt.gca())
#Data ko group karna "DateHour" ke basis par
#    aur "Sessions" aur "Users" columns ka sum nikalna
#    Fir usko plot kar dena same axes (plt.gca() = get current axes)
plt.title("Sessions and user over time views")
plt.xlabel("DateHour")
plt.ylabel("Connt")
plt.show()
```

```python
# Required libraries import
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set figure size
plt.figure(figsize=(10,5))

# Barplot: X-axis = channel group, Y-axis = Users (sum of Users per channel group)
sns.barplot(
    data=df, 
    x="channel group", 
    y="Users", 
    estimator=np.sum,   # group-wise sum calculate karega
    palette="viridis"   # color theme
)

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


# Figure size set karo
plt.figure(figsize=(8,5))

# Barplot draw karo
sns.barplot(
    data=df,
    x="channel group",
    y="Average engagement time per session",
    estimator=np.mean,   # group-wise mean calculate karega
    palette="magma"       # color theme
)

# Plot ke labels aur title
plt.title("Average Engagement Time per Session")
plt.xticks(rotation=45)


# Plot show karo
plt.show()
```
